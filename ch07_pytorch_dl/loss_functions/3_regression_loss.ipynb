```python
import torch
import torch.nn as nn
from torch.nn import SmoothL1Loss
```

```python
# 定义数据
input = torch.randn(3,5)
target = torch.randn(3,5)
print(input)
print(target)
```

```python
# MSE:L2 Loss
mse_loss = nn.MSELoss()
print(mse_loss(input, target))
```

```python
# MSE:L1 Loss
mse_loss = nn.L1Loss()
print(mse_loss(input, target))
```

```python
# Smooth L1
smooth_l1_loss = nn.SmoothL1Loss()
print(smooth_l1_loss(input, target))
```
