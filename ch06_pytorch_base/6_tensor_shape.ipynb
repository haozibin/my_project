```python
import torch
```

```python
tensor1 = torch.randint(1,10,(2,3,6))
print(tensor1)
```

6.8.1 交换维度

```python
# 1.交换维度
print(tensor1.transpose(0,2))
```

```python
# 2.重新排列多个维度
print(tensor1.permute(0,2,1))
```

6.8.2 调整形状

```python
# 1. reshape
tensor1 = torch.randint(1,10,(3,5,4))
print(tensor1)
print(tensor1.shape)
```

```python
print(tensor1.reshape(6,10))
```

```python
print(tensor1.reshape(3,-1))
```

```python
print(tensor1.reshape(-1))
```

```python
# 2.view
print(tensor1.is_contiguous())
```

```python
print(tensor1.view(-1,10))
```

```python
print(tensor1.shape)
```

```python
# 对张量进行转置操作，让内存空间不连续
tensor1 = tensor1.permute(2,1,0)
print(tensor1.shape)
```

```python
print(tensor1.is_contiguous())
```

```python
print(tensor1.contiguous().view(-1,10))
```

6.8.3 增加或删除维度

```python
# 1.增加维度
tensor1 = torch.tensor([1,2,3])
print(tensor1)
```

```python
# 在0维上增加一个维度，1*3
print(tensor1.unsqueeze(dim=0))
```

```python
# 在1维上增加一个维度，3*1
print(tensor1.unsqueeze(dim=1))
print(tensor1)
```

```python
print(tensor1.unsqueeze_(dim=-2))
print(tensor1)
```

```python
# 减少维度
print(tensor1.squeeze(dim=0))
print(tensor1)
```

6.9 张量拼接操作

```python
# 1.拼接
tensor1 = torch.randint(1,10,(2,2,5))
tensor2 = torch.randint(1,10,(2,3,5))
print(tensor1)
print(tensor2)
```

```python
print(torch.cat((tensor1,tensor2),dim=1))
```

```python
# 2. 堆叠
tensor1 = torch.randint(1,10,(4,3,5))
tensor2 = torch.randint(1,10,(4,3,5))
print(tensor1)
print(tensor2)
```

```python
tensor3 = torch.stack([tensor1,tensor2],dim=0)
print(tensor3)
print(tensor3.shape)
```

```python
tensor3 = torch.stack([tensor1,tensor2],dim=3)
print(tensor3)
print(tensor3.shape)
```
