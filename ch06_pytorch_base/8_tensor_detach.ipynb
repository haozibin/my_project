```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x.detach()
print(x)
print(y)
```

```python
print(x.requires_grad)
print(y.requires_grad)
```

```python
print(id(x))
print(id(y))
```

```python
print(x.untyped_storage().data_ptr())
print(y.untyped_storage().data_ptr())
```

```python
# 分别对x、y进行后续计算
z1 = x ** 2
z2 = y ** 2
print(z1)
print(z2)
```

```python
print(z1.requires_grad)
print(z2.requires_grad)
```

```python
z1.sum().backward()
```

```python
z2.sum().backward()
```

```python
x = torch.ones(2,2,requires_grad=True)
y = x * x
print(x)
print(y)
```

```python
# 分离y
u = y.detach()
print(u)
```

```python
# 让u参与新的张量的计算
z = u * x
```

```python
#反向传播
z.sum().backward()
```

```python
print(x.grad)
```

```python
print(x.grad == u)
```
