```python
import torch
```

```python
# 定义数据
x = torch.tensor(10.0)
y = torch.tensor([[3.0]])
print(x)
print(y)
```

```python
# 初始化参数
w = torch.rand(1,1,requires_grad=True)
b = torch.rand(1,1,requires_grad=True)
```

```python
# 前向传播，得到输出
z = w * x + b
print(z)
```

```python
print(x.is_leaf)
print(y.is_leaf)
print(w.is_leaf)
print(b.is_leaf)
print(z.is_leaf)
```

```python
#设置损失函数
loss = torch.nn.MSELoss()
loss_value = loss(z,y)
print(loss_value)
print(loss_value.is_leaf)
```

```python
# 反向传播
loss_value.backward()
```

```python
# 查看梯度
print(w.grad)
print(b.grad)
```
